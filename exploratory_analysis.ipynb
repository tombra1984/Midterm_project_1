#### **Task 7**: Test the hypothesis whether planes fly faster when there is the departure delay? 

```python
flights_df['distance'].value_counts()
```

```python
# Split the data into two groups based on departure delay
delayed_flights = flights_df[flights_df['dep_delay'] > 0]
non_delayed_flights = flights_df[flights_df['dep_delay'] <= 0]

# Remove missing values from the airtime data
delayed_airtime = delayed_flights['air_time'].dropna()
non_delayed_airtime = non_delayed_flights['air_time'].dropna()

# Calculate the mean airtime for each group
mean_airtime_delayed = np.mean(delayed_airtime)
mean_airtime_non_delayed = np.mean(non_delayed_airtime)

# Perform a hypothesis test to compare the means of the two groups
_, p_value = stats.ttest_ind(delayed_airtime, non_delayed_airtime, alternative='greater')

print(f"Mean airtime for delayed flights: {mean_airtime_delayed:.2f} minutes")
print(f"Mean airtime for non-delayed flights: {mean_airtime_non_delayed:.2f} minutes")
print(f"P-value: {p_value:.4f}")

## This does not consider flights with same distances, therefore the air-time mean may vary.
```

```python
YOUR_DISTANCE = 337

# Filter flights with departure delay
delayed_flights = flights_df[flights_df['dep_delay'] > 0]

# Filter flights without departure delay
non_delayed_flights = flights_df[flights_df['dep_delay'] <= 0]

# Remove missing values from airtime data
delayed_airtime = delayed_flights['air_time'].dropna()
non_delayed_airtime = non_delayed_flights['air_time'].dropna()

# Filter flights with the same distance
same_distance_delayed = delayed_flights[delayed_flights['distance'] == YOUR_DISTANCE]
same_distance_non_delayed = non_delayed_flights[non_delayed_flights['distance'] == YOUR_DISTANCE]

# Remove missing values from airtime data for flights with the same distance
same_distance_delayed_airtime = same_distance_delayed['air_time'].dropna()
same_distance_non_delayed_airtime = same_distance_non_delayed['air_time'].dropna()

# Calculate the mean airtime for each group
mean_airtime_delayed = np.mean(same_distance_delayed_airtime)
mean_airtime_non_delayed = np.mean(same_distance_non_delayed_airtime)

# Perform a hypothesis test to compare the means of the two groups
_, p_value = stats.ttest_ind(same_distance_delayed_airtime, same_distance_non_delayed_airtime, alternative='greater')

print(f"Mean airtime for delayed flights with distance {YOUR_DISTANCE}: {mean_airtime_delayed:.2f} minutes")
print(f"Mean airtime for non-delayed flights with distance {YOUR_DISTANCE}: {mean_airtime_non_delayed:.2f} minutes")
print(f"P-value: {p_value:.4f}")

## This considers flights with a distance of 337, which we used because this distance is the most commonly travelled route.
```

```python
## Both tests showed that the hypothesis will be rejected. 
```

#### **Task 8**: When (which hour) do most 'LONG', 'SHORT', 'MEDIUM' haul flights take off?

```python
# Define the distance thresholds for flight categories
short_threshold = 500
medium_threshold = 1500

# Categorize flights based on distance
flights_df['flight_distance_category'] = pd.cut(flights_df['distance'], bins=[0, short_threshold, medium_threshold, np.inf], labels=['SHORT', 'MEDIUM', 'LONG'])

# Extract the hour from the scheduled departure time
flights_df['dep_hour'] = flights_df['crs_dep_time'] // 100

# Group the data by flight distance category and departure hour, and count the number of flights
flight_hour_counts = flights_df.groupby(['flight_distance_category', 'dep_hour']).size().reset_index(name='count')

# Find the hour(s) with the highest count for each flight distance category
max_hour_short = flight_hour_counts.loc[flight_hour_counts[flight_hour_counts['flight_distance_category'] == 'SHORT']['count'].idxmax(), 'dep_hour']
max_hour_medium = flight_hour_counts.loc[flight_hour_counts[flight_hour_counts['flight_distance_category'] == 'MEDIUM']['count'].idxmax(), 'dep_hour']
max_hour_long = flight_hour_counts.loc[flight_hour_counts[flight_hour_counts['flight_distance_category'] == 'LONG']['count'].idxmax(), 'dep_hour']

# Print the results
print("Hour(s) with the most SHORT haul flights:", max_hour_short)
print("Hour(s) with the most MEDIUM haul flights:", max_hour_medium)
print("Hour(s) with the most LONG haul flights:", max_hour_long)

```

```python

# Group the flights by flight_distance_category and dep_hour, and count the number of flights
grouped_flights = flights_df.groupby(['flight_distance_category', 'dep_hour']).size().unstack()

# Plot the bar plot
fig, ax = plt.subplots(figsize=(10, 6))
grouped_flights.plot(kind='bar', ax=ax)

# Set the plot title and labels
ax.set_title("Distribution of Flights by Flight Distance and Departure Hour")
ax.set_xlabel("Departure Hour")
ax.set_ylabel("Number of Flights")

# Show the legend
ax.legend(title='Hour of the day')

# Display the plot
plt.show()
```

```python
# Step 1: Calculate the average monthly departure delay per air carrier
average_departure_delay = flights_df.groupby(['mkt_unique_carrier', 'month'])['dep_delay'].mean().reset_index()

# Step 2: Calculate the total distance covered monthly by different air carriers
total_distance_covered = flights_df.groupby(['mkt_unique_carrier', 'month'])['distance'].sum().reset_index()

# Step 3: Calculate the total fuel consumption per air carrier
total_fuel_consumption = fuel_consumption.groupby(['unique_carrier', 'year'])['total_gallons'].sum().reset_index()

# Merge the dataframes to calculate fuel consumption per mile
merged_data = pd.merge(average_departure_delay, total_distance_covered, on=['mkt_unique_carrier', 'month'])
merged_data = pd.merge(merged_data, total_fuel_consumption, left_on='mkt_unique_carrier', right_on='unique_carrier')

# Calculate fuel consumption per mile
merged_data['fuel_consumption_per_mile'] = merged_data['total_gallons'] / merged_data['distance']

# Compare average departure delays with fuel consumption per mile
merged_data['average_departure_delay'] = merged_data.groupby('mkt_unique_carrier')['dep_delay'].transform('mean')

# Separate airlines with higher average departure delays and lower average departure delays
higher_delay_airlines = merged_data[merged_data['average_departure_delay'] > merged_data['average_departure_delay'].mean()]
lower_delay_airlines = merged_data[merged_data['average_departure_delay'] <= merged_data['average_departure_delay'].mean()]

# Calculate the average fuel consumption per mile for each group
average_fuel_consumption_higher_delay = higher_delay_airlines['fuel_consumption_per_mile'].mean()
average_fuel_consumption_lower_delay = lower_delay_airlines['fuel_consumption_per_mile'].mean()

# Print the results
print("Average fuel consumption per mile for airlines with higher average departure delays:", average_fuel_consumption_higher_delay)
print("Average fuel consumption per mile for airlines with lower average departure delays:", average_fuel_consumption_lower_delay)
```
